# 開花日期資料

```python
# 讀取原始資料
blossom_date = pd.read_csv("Dataset/sakura_blossom_date_from_1961_to_2024.csv")

# 直接針對「開花日」欄位，把最後出現的 "3 24" 移除
blossom_date["開花日"] = blossom_date["開花日"].apply(
    lambda x: str(x).rsplit("3 24", 1)[0].strip()
)
blossom_date
```

```python
# 初始化結果儲存列表
all_years = []
all_months = []
all_days = []

# 處理每一列（每列是一段年份區間）
for idx, row in blossom_date.iterrows():
    # 解析年份區間，例如 "1961–1970"
    year_range = row.iloc[0]
    try:
        start_year, end_year = map(int, year_range.split("–"))
    except:
        continue

    year_count = end_year - start_year + 1

    # 擷取日期文字欄位並清理
    raw_string = str(row.iloc[1])
    cleaned = raw_string.replace("東京", "").replace("*", "")
    tokens = cleaned.split()
    tokens = [t for t in tokens if t.isdigit()]

    # 拆解成 (月, 日) 對
    raw_date_pairs = []
    for i in range(0, len(tokens) - 1, 2):
        try:
            month = int(tokens[i])
            day = int(tokens[i + 1])
            raw_date_pairs.append((month, day))
        except:
            continue

    # 只保留最後 year_count 筆資料（避免前面資料錯誤）
    date_pairs = raw_date_pairs[-year_count:]

    # 加入對應年份
    for i, (month, day) in enumerate(date_pairs):
        year = start_year + i
        all_years.append(year)
        all_months.append(month)
        all_days.append(day)

# 建立整理後的 DataFrame
blossom_date = pd.DataFrame({
    "年": all_years,
    "月": all_months,
    "日": all_days
})

blossom_date
```

```python
# 建立從1961-01-01到2024-12-31的完整日期範圍
full_date_range = pd.date_range(start="1961-01-01", end="2024-12-31")

# 將日期範圍轉換為 DataFrame，拆分成年、月、日欄位
blossom_date_expanded = pd.DataFrame({
    "年": full_date_range.year,
    "月": full_date_range.month,
    "日": full_date_range.day
})

# 新增一欄 blossom_date，預設為 0（表示非開花日）
blossom_date_expanded["開花日"] = 0

# 將原始開花日資料新增一欄 "date"，轉為 datetime 格式以便比對
blossom_date["日期"] = pd.to_datetime(dict(year=blossom_date["年"], 
                                                     month=blossom_date["月"], 
                                                     day=blossom_date["日"]
                                                    )
                                               )

# 將完整資料中的日期也轉為 datetime 格式
blossom_date_expanded["日期"] = pd.to_datetime(dict(year=blossom_date_expanded["年"], 
                                                    month=blossom_date_expanded["月"], 
                                                    day=blossom_date_expanded["日"]
                                                   )
                                              )

# 比對日期，若是開花日，將 blossom_date 設為 1
blossom_date_expanded.loc[blossom_date_expanded["日期"].isin(blossom_date["日期"]), "開花日"] = 1

# 移除中間用來比對的 date 欄位
blossom_date_expanded.drop(columns="日期", inplace=True)

# 顯示前幾列查看結果
print(blossom_date_expanded.head())
```

```python
# 合併資料：依據 年、月、日 對齊
tokyo_weather_processed_merged = pd.merge(tokyo_weather, blossom_date_expanded[["年", "月", "日", "開花日"]], on=["年", "月", "日"], how="left")

# 將 "開花日" 欄位移動到最後一欄
cols = list(tokyo_weather_processed_merged.columns)
cols.remove("開花日")
cols.append("開花日")
tokyo_weather_processed_merged = tokyo_weather_processed_merged[cols]

# 顯示前幾筆資料確認
print(tokyo_weather_processed_merged.head())
```

### 按照年份順序切 8:2

```python
# 計算 80% 的切割點
split_index = int(len(all_years) * 0.8)

# 分成前 80% 年份為訓練集，後 20% 年份為測試集
train_years = all_years[:split_index]
test_years = all_years[split_index:]

# 根據年份分割資料
train = tokyo_weather_processed_merged[tokyo_weather_processed_merged["年"].isin(train_years)]
test = tokyo_weather_processed_merged[tokyo_weather_processed_merged["年"].isin(test_years)]

# 檢查分割結果
print("訓練集年份：", train_years[0], "~", train_years[-1])
print("測試集年份：", test_years[0], "~", test_years[-1])

# 可選：將分割後的資料存成 CSV
train.to_csv("Dataset/train.csv", index=False)
test.to_csv("Dataset/test.csv", index=False)

print("CSV 檔案已成功儲存成功")
```
