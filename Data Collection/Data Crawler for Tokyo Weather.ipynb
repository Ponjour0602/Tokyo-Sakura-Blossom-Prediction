```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from io import StringIO
```

### ็ฌๅๆฅๆฌๅๅไบค้็ๆฐฃ่ฑกๅปณ็ๆฐฃ่ฑก่ณๆ
### ไปฅไธ้ฃ็ต็บ"2024"ๅนด"1"ๆ"ๆฑไบฌ"็ๆฐฃ่ฑกไฝ็บไพๅญ
#### https://www.data.jma.go.jp/stats/etrn/view/daily_s1.php?prec_no=44&block_no=47662&year=2024&month=1&day=&view=

```python
# ็ฎๆจ็ถฒๅ๏ผ่ณๆ้กฏ็คบ้๏ผ
base_url = "https://www.data.jma.go.jp/stats/etrn/view/daily_s1.php"
# ๆฅ่ฉขๅๆธ็ฏๆฌ๏ผๅฐ้ป๏ผๆฑไบฌ๏ผ
params = {
    "prec_no": "44",       # ๆฑไบฌ็้ฝ้ๅบ็ธฃ็ทจ่
    "block_no": "47662",   # ๆฑไบฌ็ๆฐฃ่ฑก็ซ็ทจ่
    "year": "",            # ๅนดไปฝ๏ผๅพ้ขๅกซ๏ผ
    "month": "",           # ๆไปฝ๏ผๅพ้ขๅกซ๏ผ
    "day": "",             # ็็ฉบไปฃ่กจๆดๆ
    "view": ""             # ้่จญๆชข่ฆ
}

# ๅปบ็ซ็ฉบ็ DataFrame ๅฒๅญๆๆ่ณๆ
all_data = None
column_names = None

# ่ฟดๅ่ทๆฏๅนดใๆฏๆ
for year in range(1961, 2025):  # 1961 ๅฐ 2024๏ผๅซ๏ผ
    for month in range(1, 13):  # ๆฏๅนด 1 ๅฐ 12 ๆ
        print(f"๐ฆ ๆๅ {year} ๅนด {month} ๆ...")
        params["year"] = str(year)
        params["month"] = str(month)
        try:
            # ็ผ้ GET ่ซๆฑ
            response = requests.get(base_url, params=params)
            if response.status_code != 200:
                print(f"โ ็กๆณๅๅพ {year}/{month} ็่ณๆ๏ผHTTP ้ฏ่ชค็ขผ๏ผ{response.status_code}")
                continue
                
            # ็จ BeautifulSoup ่งฃๆ HTML
            soup = BeautifulSoup(response.content, "html.parser")
            table = soup.find("table", class_="data2_s")
            
            if table:
                # ็จ StringIO ้ฟๅ่ญฆๅ
                df = pd.read_html(StringIO(str(table)), header=0)[0]
                
                # ๅฆๆๆฏ็ฌฌไธๆฌก่ฎๅ๏ผไฟๅญcolumnๅ็จฑ
                if all_data is None:
                    column_names = df.columns
                    all_data = df.copy()
                else:
                    # ไฝฟ็จ็ฌฌไธๆฌก็ฒๅ็columnๅ็จฑ
                    df.columns = column_names
                    # ๅป้คๅฏ่ฝ็ๆจ้ก่ก (ๅฆๆๆจ้ก่ก่ฎๆไบๆธๆ)
                    df = df[~df.iloc[:, 0].astype(str).str.contains("ๆฅ")]
                    all_data = pd.concat([all_data, df], ignore_index=True)
                
                # ๅไธๅนดไปฝ่ๆไปฝๆฌไฝ
                all_data.loc[all_data.shape[0]-df.shape[0]:, "ๅนด"] = year
                all_data.loc[all_data.shape[0]-df.shape[0]:, "ๆ"] = month
                
            else:
                print(f"โ๏ธ ๆพไธๅฐ่กจๆผ๏ผ{year} ๅนด {month} ๆ")
                
            # ้จๆฉไผๆฏ 1~3 ็ง๏ผ้ฟๅ่ขซๅฐ้
            sleep(randint(1, 3))
            
        except Exception as e:
            print(f"โ๏ธ ้ฏ่ชค๏ผ{year}/{month} ็ผ็็ฐๅธธ๏ผ{e}")
            continue

# ๅฒๅญๆ CSV๏ผๅฏๆนๆ Excel๏ผ
all_data.to_csv("tokyo_weather_from_1961_to_2024.csv", index=False, encoding="utf-8-sig")
print("โ ๆๆ่ณๆๅทฒๅฎๆไธ่ผ๏ผๅฒๅญ็บ tokyo_weather_from_1961_to_2024.csv")
```
